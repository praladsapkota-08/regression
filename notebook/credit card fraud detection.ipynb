```python
from sklearn.preprocessing import StandardScaler

cc['Amount'] = np.log1p(df['Amount'])

sns.kdeplot(cc['Amount'])
scaler = StandardScaler()

cc['scaled amount'] = scaler.fit_transform(cc[['Amount']])
cc['scaled hours'] = scaler.fit_transform(cc[['hours']])

cc
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

x = cc.drop(columns = ['Class'] , axis = 1)
y = cc['Class']

x_train,x_test,y_train,y_test = train_test_split(x , y , test_size = 0.2, random_state = 42)
smote = SMOTE(random_state = 42)
x_train_sample ,y_train_sample = smote.fit_resample(x_train,y_train)

print('original dataset', y_train.value_counts())
print('sampled dataset',y_train_sample.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = cc.drop(columns = ['Class'] , axis = 1)
y = cc['Class']

x_train,x_test,y_train,y_test = train_test_split(x , y , test_size = 0.2, random_state = 42)
log_model = LogisticRegression(random_state = 42)

log_model.fit(x_train_sample, y_train_sample)
```

```python
y_pred = log_model.predict(x_test)
y_probab = log_model.predict_proba(x_test)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix , classification_report , average_precision_score

print('Confusion metrix' , confusion_matrix(y_test,y_pred))
print('classification report', classification_report(y_test,y_pred))
auprc = average_precision_score(y_test,y_probab)
print(f'AUPRC score {auprc:.4f}')
```

```python
neg , pos = np.bincount(y_train)

scale_pos_weight = neg / pos
print(f"scale pos weight {scale_pos_weight}")
```

```python
from xgboost import XGBClassifier

# model = XGBClassifier(
#     scale_pos_weight=scale_pos_weight,
#     n_estimators=100,
#     learning_rate=0.1,
#     max_depth=4,
#     random_state=42,
#     use_label_encoder=False,
#     eval_metric='logloss'
# )

# model.fit(X_train, y_train)

xgb_class_no_smote = XGBClassifier(
    scale_pos_weight = scale_pos_weight,
    n_estimators = 100,
    learning_rate = 0.1,
    max_depth = 4,
    random_state = 42,
    use_label_encoder = False,
    eval_metric = 'logloss'
)

xgb_class_no_smote.fit(x_train,y_train)

print(xgb_class_no_smote)
```

```python
y_pred_2 = xgb_class_no_smote.predict(x_test)
y_probab_2 = xgb_class_no_smote.predict_proba(x_test)[:,1]
print('confusion matrix', confusion_matrix(y_test,y_pred_2))
print('classification report' , classification_report(y_test, y_pred_2))

auprc = average_precision_score(y_test,y_probab_2)

print(f'auprc {auprc}')
```

```python
xgb_class_smote = XGBClassifier(
    n_estimators = 100,
    learning_rate = 0.1,
    max_depth = 4,
    random_state = 42,
    use_label_encoder = False,
    eval_metric = 'logloss'
    )

xgb_class_smote.fit(x_train_sample,y_train_sample)

print(xgb_class_smote)
```
